```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_movies3.csv')
```

```python
df
```

```python
# update the budget with 2019 dollar index
cpi_dict = {2010: 218.056, 2011: 224.939, 2012: 229.594, 2013: 232.957, 2014: 236.736, 2015: 237.017, 2016: 240.007, 2017: 245.120, 2018: 251.107, 2019: 255.657}

# target CPI (year 2019)
target_cpi = cpi_dict[2019]

def adjust_budget_to_2019(budget, year, cpi_dict, target_cpi):
    year_cpi = cpi_dict.get(year) 
    adjusted_budget = (budget * target_cpi) / year_cpi
    return round(adjusted_budget, 2)

# Yeni sütun ekleme
df["Adjusted_to_2019_BoxOffice"] = df.apply(
    lambda row: adjust_budget_to_2019(row["Budget"], row["Year"], cpi_dict, target_cpi), axis=1
)
```

```python
df
```

```python
df.drop(columns=['Budget'], inplace=True)
df.rename(columns={'Adjusted_to_2019_BoxOffice': 'Budget'}, inplace=True)
```

```python
# drop string min from runtime 
df['Runtime'] = df['Runtime'].str.replace('min', '').astype(int)

# and convert all values of runtime to int
df['Runtime'] = df['Runtime'].astype(int)

# change column name as Runtime(min)
df.rename(columns={'Runtime': 'Runtime(min)'}, inplace=True)
```

```python
df.to_csv('cleaned_movies4.csv', index=False)
```
