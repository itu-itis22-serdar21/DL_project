```python
# Clean the data by removing rows with missing values
df = df.dropna()
```

```python
# how many films are there short films  as genre(it can contain short films and other genres)
df['Genre'].str.contains('Short').sum()
```

```python
# remove the short films from dataset
df = df[~df['Genre'].str.contains('Short')]
```

```python
# Remove the years that are not integers
df = df[df['Year'].str.isnumeric()]
```

```python
# Take the first country in the list as a country
df['Country'] = df['Country'].str.split(',').str[0]
```

```python
value_counts = df['Country'].value_counts()
print(value_counts)
```

```python
# get top 2 values of index
vals = value_counts[:8].index
print (vals)
df['Country'] = df.Country.where(df.Country.isin(vals), 'Other')
```

```python
# successfully divided the country into three catogories 
df['Country'].value_counts()
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df['BoxOffice'] = df['BoxOffice'].str.replace('$', '')
df['BoxOffice'] = df['BoxOffice'].str.replace(',', '')
df['BoxOffice'] = df['BoxOffice'].astype(int)
```

```python
# update the budget with 2019 dollar index
cpi_dict = {2010: 218.056, 2011: 224.939, 2012: 229.594, 2013: 232.957, 2014: 236.736, 2015: 237.017, 2016: 240.007, 2017: 245.120, 2018: 251.107, 2019: 255.657}

# target CPI (year 2019)
target_cpi = cpi_dict[2019]

def adjust_budget_to_2019(budget, year, cpi_dict, target_cpi):
    year_cpi = cpi_dict.get(year) 
    adjusted_budget = (budget * target_cpi) / year_cpi
    return round(adjusted_budget, 2)

# Yeni sütun ekleme
df["Adjusted_to_2019_BoxOffice"] = df.apply(
    lambda row: adjust_budget_to_2019(row["BoxOffice"], row["Year"], cpi_dict, target_cpi), axis=1
)
```

```python
# are there any duplicate rows
df.duplicated().sum()
```

```python
# seperate the actors and directors into different columns (actor1, actor2 etc.) range is not fixed, it depends on the movie
df['Actors'] = df['Actors'].str.split(', ')
df['Director'] = df['Director'].str.split(', ')
```

```python
# find the maximum number of actors and directors in a movie
max_actors = df['Actors'].apply(len).max()
max_directors = df['Director'].apply(len).max()
```

```python
# find minimum number of actors and directors
min_actors = df['Actors'].apply(len).min()
min_directors = df['Director'].apply(len).min()
```

```python
print(max_actors, max_directors)
```

```python
print(min_actors, min_directors)
```

```python
# actor1, actor2 actor3, director1
df['Actor1'] = df['Actors'].str[0]
df['Actor2'] = df['Actors'].str[1]
df['Actor3'] = df['Actors'].str[2]
df['Director'] = df['Director'].str[0]
```

```python
df.drop(columns=['Actors'], inplace=True)
```

```python
# seperate the genres into different columns
df['Genre'] = df['Genre'].str.split(', ')
```

```python
# find minimum number of actors and directors
min_genre = df['Genre'].apply(len).min()
max_genre = df['Genre'].apply(len).max()
```

```python
print(min_genre, max_genre)
```

```python
# genre1, genre2 genre3 if there is no genre, fill it with genre1

# Ensure Genre1 is always the first genre or NaN if Genre is empty
df['Genre1'] = df['Genre'].str[0]

# Fill Genre2 and Genre3 with appropriate values or fall back to Genre1
df['Genre2'] = df['Genre'].str[1].fillna(df['Genre1'])
df['Genre3'] = df['Genre'].str[2].fillna(df['Genre1'])
```

```python
df.drop(columns=['Genre'], inplace=True)
```

```python
df.drop(columns=['BoxOffice'], inplace=True)
```

```python
# rename the adjust_to_2019_boxoffice column to boxoffice
df.rename(columns={'Adjusted_to_2019_BoxOffice': 'BoxOffice'}, inplace=True)
```

```python
df.drop(columns=['Poster'], inplace=True)
```

```python
# Save the cleaned data to a new csv file
df.to_csv('cleaned_movies.csv', index=False)
```
